```python
import torch
import numpy as np

fake_image = torch.zeros(1,3,640,640)
print(f"shape = {fake_image.shape}")
print(f"data type = {fake_image.dtype}")
fake_image_normalised = fake_image/255.0
print(f"After normalise — min: {fake_image_normalised.min()}, max: {fake_image_normalised.max()}")

H = torch.tensor([
    [ 0.1,  0.0, -5.0],
    [ 0.0,  0.1, -3.0],
    [ 0.0,  0.0,  1.0]
], dtype=torch.float64)
print(f"\nHomography shape: {H.shape}")

pixel = torch.tensor([[320.0], [240.0], [1.0]], dtype=torch.float64)
print(f"\nDetected pixel: {pixel.T}")

result = H @ pixel
result = result / result[2]
print(f"Real world position: X={result[0].item():.1f}mm, Y={result[1].item():.1f}mm")
```
